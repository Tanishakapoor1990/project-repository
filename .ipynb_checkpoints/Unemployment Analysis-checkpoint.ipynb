### Unemployment Rate by Race

```python
import pandas as pd
import requests
```

```python
API_KEY = 'tT8qNAwBZW89z54wPsZy'
url = 'https://www.quandl.com/api/v3/datasets/FRED/LNS14000003.json?api_key='+ API_KEY
response = requests.get(url).json()
response
```

```python
name_list = response['dataset']['name']
data_list = response['dataset']['data']
data_list
unemploymentrate_white = pd.DataFrame(data_list,columns=['date' , name_list]) # converting list to df
unemploymentrate_white
```

```python
def my_func(a= 'LNS14000006'):
    
    API_KEY = 'tT8qNAwBZW89z54wPsZy'
    url = 'https://www.quandl.com/api/v3/datasets/FRED/'+a+'.json?api+key='+API_KEY
    response = requests.get(url).json()
    response
#"https://www.quandl.com/api/v3/datasets/FRED/LNS14000003/
    name = response['dataset']['name']
    data = response['dataset']['data']
    df = pd.DataFrame(data,columns=['date' , name])
    return df
```

```python
unemploymentrate_hispanic = my_func(a = 'LNS14000009')
unemploymentrate_hispanic
unemploymentrate_asian = my_func(a= 'LNU04032183')
unemploymentrate_asian
unemploymentrate_african = my_func(a='LNS14000006')
unemploymentrate_african
unemployment_rate = my_func(a = 'UNRATE')
unemployment_rate
```

```python
unemploymentrate_hispanic
unemploymentrate_asian
#data2
#data1
#data2
```

```python
modified_df  = pd.merge(unemploymentrate_hispanic,unemploymentrate_asian,how='outer', on='date')
modified_df
```

```python
df3=pd.merge(modified_df,unemploymentrate_african, how='outer', on='date')
unemploymentrate_race=pd.merge(df3,unemploymentrate_white,how='outer', on='date')
unemploymentrate_race
unemploymentrate_race_df=pd.merge(unemploymentrate_race,unemployment_rate,how='outer', on='date')
unemploymentrate_race_df
```

### Unemployment Rate by Gender 

```python
def my_func(a= 'LNS14000001'):
    
    API_KEY = 'tT8qNAwBZW89z54wPsZy'
    url = 'https://www.quandl.com/api/v3/datasets/FRED/'+a+'.json?api+key='+API_KEY
    response = requests.get(url).json()
    response
#"https://www.quandl.com/api/v3/datasets/FRED/LNS14000003/
    name = response['dataset']['name']
    data = response['dataset']['data']
    df = pd.DataFrame(data,columns=['date' , name])
    return df
```

```python
unemploymentrate_men =my_func(a= 'LNS14000001')
unemploymentrate_men 
unemploymentrate_women= my_func(a= 'LNS14000002')
unemploymentrate_women
```

```python
unemploymentrate_gender= pd.merge(unemploymentrate_men,unemploymentrate_women,how='outer',on='date')
unemploymentrate_gender
```

```python
unemploymentrate_race_gender_df=pd.merge(unemploymentrate_race_df,unemploymentrate_gender,how='outer',on='date')
unemploymentrate_race_gender_df
```

## For converting date to datetime data type

```python
unemploymentrate_race_gender_df.dtypes
```

```python
unemploymentrate_race_gender_df["date"]=pd.to_datetime(unemploymentrate_race_gender_df["date"])
unemploymentrate_race_gender_df.dtypes
```

```python
unemploymentrate_race_gender_df.columns
```

## Renaming column names

```python
unemploymentrate_race_gender_df.rename(columns ={'date': 'Date','Unemployment Rate: Hispanic or Latino' :'Unemp_Hispanic',
                                                 'Unemployment Rate: Asian':'Unemp_Asian',
                                                 'Unemployment Rate: Black or African American':'Unemp_Black',
                                                'Unemployment Rate: White':'Unemp_White',
                                               'Civilian Unemployment Rate' :'Unemp_Civilian','Unemployment Rate: Men' :'Unemp_Men',
                                               'Unemployment Rate: Women':'Unemp_Women'},inplace=True)
unemploymentrate_race_gender_df
```

```python
unemploymentrate_race_gender_df
```

```python
unemploymentrate_race_gender_df["Unemp_White_Men"] =unemploymentrate_race_gender_df["Unemp_White"] * unemploymentrate_race_gender_df["Unemp_Men"]/1000*100
unemploymentrate_race_gender_df
```

```python
unemploymentrate_race_gender_df
```
