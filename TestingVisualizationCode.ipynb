```python
# import dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set()
```

```python
df = pd.read_csv("unemp_rate.csv")
#df.drop(df.index[1967-12-1]:[1948-1-1], inplace = True)
```

```python
#checking out subset of data before dropping rows
start_date = "2018-09-01"
end_date = "2021-03-01"


mask = (df['Date'] > start_date) & (df['Date'] <= end_date)
df = df.loc[mask]
df
```

```python
df.drop(df.index[639:879], inplace=True)
df.sort_values(by=['Date'], inplace = True)
df.set_index('Date', inplace = True)
df.head()
```

```python
df[['20-24','25-34','35-44', '45-54', '55 and up', 'Unemp_Civilian']].plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
df_COVID = pd.read_csv("unemp_rate.csv")

#checking out subset of data before dropping rows
#start_date = "2018-09-01"
#end_date = "2021-03-01"


#mask = (df_COVID['Date'] > start_date) & (df_COVID['Date'] <= end_date)
#df_COVID = df_COVID.loc[mask]
#df_COVID

df_COVID.drop(df_COVID.index[30:879], inplace=True)
df_COVID.sort_values(by=['Date'], inplace = True)
df_COVID.set_index('Date', inplace = True)
df_COVID
```

```python
df_COVID[['Unemp_Hispanic','Unemp_Asian','Unemp_Black', 'Unemp_White','Unemp_Civilian']].plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

As you can see, once we disaggregate the unemployment data according to race or ethnic group, we see that, after COVID, Hispanics sustained the highest level of unemployment (initially), followed by Blacks, then Asians, and Whites. By June of 2020, the highest rate of unemployment was for Blacks, followed by Hispanics and Asians (close). By the end of 2020, unemployment rate was lowest for Whites and Asians and still appreciably higher for Blacks and Hispanics.
